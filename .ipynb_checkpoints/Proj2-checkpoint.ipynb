```python
import pandas as pd
import os
from tqdm.auto import tqdm
import matplotlib.pyplot as plt
import json
import numpy as np
from collections import OrderedDict
from collections import Counter
pd.options.mode.chained_assignment = None  # default='warn
```

```python
############################ FUNCTIONS ########################################
def loadData():
    cwd = os.getcwd()
    df = pd.read_csv('./ml-latest-small/movies.csv')
    genre_df = df['genres'].apply(lambda x: x.split("|"))
    users = pd.read_csv('./ml-latest-small/ratings.csv')

    return df, genre_df, users

def query_genre():
    wanted_genre = 'NULL'
    uniques = []
    for x in genre_df:
        if x:
            for y in x:
                if y.lower() not in uniques:
                    uniques.append(y.lower())

    print("________________________***GENRES***_______________________________")
    i = 0
    next = 2
    for x in range(len(uniques)):
        if next+2 <= len(uniques):
            if(not i):
                print("{:30s} {:2s}".format(str(uniques[i]) , str(uniques[i+1])))
            else:
                print("{:30s} {:2s}".format(str(uniques[next]) , str(uniques[next+1])))
                next+=2
        i+=1

    print("________________________***GENRES***_______________________________\n")

    while(wanted_genre.lower() not in uniques):
        wanted_genre = input('What genre do you feel like watching? ')
        if wanted_genre not in uniques:
            print("Please Try Again...\n")

    suggestions = []

    result = df[df['genres'].str.contains(wanted_genre, case=False, na=False)]

    if result.empty:
        print("No results for " + wanted_genre)
        return False
    else:
        print("_________________**RECOMMENDATIONS**___________________________")

        i = 0
        next = 2
        for x in range(len(result['title'].values)):
            if next+2 <= len(result['title'].values):
                if(not i):
                    text1 = (result['title'].values[i][:75] + '..') if len(result['title'].values[i]) > 75 else result['title'].values[i]
                    text2 = (result['title'].values[i+1][:75] + '..') if len(result['title'].values[i+1]) > 75 else result['title'].values[i+1]
                    print("{:80s} {:5s}".format(text1 , text2))
                else:
                    text1 = (result['title'].values[next][:75] + '..') if len(result['title'].values[next]) > 75 else result['title'].values[next]
                    text2 = (result['title'].values[next+1][:75] + '..') if len(result['title'].values[next+1]) > 75 else result['title'].values[next+1]
                    print("{:80s} {:5s}".format(text1 , text2))
                    next+=2
            i+=1

        print("_________________**RECOMMENDATIONS**___________________________\n")
        print()

    return True

def top_genres(user):
    top = {}
    for movie in user['genres']:
        gens = movie.split('|')
        for genre in gens:
            if genre not in top.keys():
                top[genre] = 1
            else:
                top[genre] += 1

    top_sorted = dict(sorted(top.items(), key=lambda item: item[1]))
    return list(top_sorted.keys())

def show_genres(res):
    print("TOP " + str(len(res)) + " GENRES:\n")
    for e in res:
        print(e)

def avgRate(data):
    print('Acquiring Average Ratings of Movies...')
    res = []
    r = {}
    cnt = data['movieId'].unique().shape[0]
    with tqdm(range(data.shape[0])) as pbar:
        for id in data['movieId'].unique():
            r[id] = data[data['movieId'] == id].groupby('movieId')['rating'].mean()
            pbar.update(5)
        for id in r.keys():
            l = data[data['movieId'] == id].count().iloc[0]
            res.extend([r[id].values] * l)
            pbar.update(5)
            
        for i in range(len(res)):
            res[i] = res[i][0]

    data['avgRating'] = res

    return data

def getWeights(data):
    w = []
    m = int(data.groupby('movieId').count().quantile(0.75)[0])
    c = np.sum(data['avgRating'].unique())/data['avgRating'].unique().shape[0]

    count = data.groupby('movieId').count()
    
    print("Calculating Weighted Ratings...")
    for i in tqdm(range(data.shape[0])):
        item = data.iloc[i,:]
        
        r = item['avgRating']  
        v = count.loc[item['movieId']]['userId']
        
        w.append(((r*v)+(c*m))/(v+m))

        
    data['weighted_rating'] = w
    return data

def movieComp(movie, df):
    df['len']      = df['genres'].str.len()
    data           = df.sort_values(by='movieId', ascending=True).drop(columns='len')
    rec_embed      = {}
    orig_embed     = {}
    for i in range(0,data.shape[0]):
        genres = data['genres'][i].split('|')
        for genre in genres:
            if genre not in orig_embed.keys() and genre not in rec_embed.keys():
                orig_embed[genre] = 0
                rec_embed[genre]  = 0
    
    recs = data[data['movieId']==movie]
    
    if '|' in recs['genres'].iloc[0]:
        rec = recs['genres'].iloc[0].split('|')
        for e in rec:
            rec_embed[e] = 1
    else:
        rec_embed[recs['genres'].iloc[0]] = 1
        
    rec_score = 0
    for key in rec_embed.keys():
        if rec_embed[key] == 1:
            rec_score += 1
    
    fin   = []
    total = []
    comp  = orig_embed

    for i in range(0,data.shape[0]):
        score  = 0
        comp = dict.fromkeys(comp, 0)
        genres = data['genres'].iloc[i].split('|')
        
        for val in genres:
            comp[val] = 1
            
        fin.append(comp)
        for key in orig_embed.keys():
            if comp[key] == 1 and rec_embed[key] == 1:
                score += 1      
        if score:      
            weighted_score = abs(rec_score - score)
        else:
            weighted_score = len(rec_embed.keys())
            
        
        total.append(weighted_score)
    
    done = pd.DataFrame(fin).fillna(0).astype('int32')     
    done['similarity'] = total
    return data.join(done)

def userRecs(df, users, user):
    frame = pd.DataFrame()
    u = users[users['userId']==user].sort_values(by='rating').iloc[::-1].iloc[:10]
    for i in range(0, u.shape[0]):
        d = df[df['movieId'] == u['movieId'].iloc[i]]
        frame = pd.concat([frame, d], axis=0, ignore_index=True)
      
    ret = pd.DataFrame()
    for i in range(0, frame.shape[0]):
        movie = frame['movieId'].iloc[i]
        ret = pd.concat([ret, movieComp(movie, df).sort_values(by='similarity').iloc[:3]], axis=0, ignore_index=True) 
     
    return ret.drop_duplicates(subset=['movieId'])
    
    
    
```

### Show recommendations based on user experience

```python
'''
Gather top similar movies to what user has watched and retrn recommendations
'''

global pbar

top_n_movies = 20
df, genre_df, users = loadData()

user = 5

ret = userRecs(df, users, user).sort_values(by='similarity').iloc[:top_n_movies]
ret.style.set_caption("Top " + str(top_n_movies) + " Movies Recommended for User: " + str(user))
```

### Show top similar movies based on given ID

```python
'''
Can change movie ID in variable 'movie'
Can chage n closest movies as well

        Similarity calculated by subtracting the value of the score of genres in the movie 
        you selected over with the score of all other movies
            
            = |wanted_score - movie_score|
'''


global pbar

n_closest_movies = 10
movie = 6

df, genre_df, users = loadData()


df = movieComp(movie, df)
result = df.sort_values(by='similarity')
result = result.iloc[:n_closest_movies]
result.style.set_caption("Top " + str(n_closest_movies) + " Movies Similar to: " + str(df[df['movieId']==movie]['title'].iloc[0]))
```

### Display top movies watched by users using IMDB weighted result calculations

```python
global pbar

top_n_movies = 20
df, genre_df, users = loadData()

users = avgRate(users.sort_values(by=['movieId']))
users = getWeights(users)
print("From gathered user data, the top " + str(top_n_movies) + " rated movies are:")
print()
res = users.sort_values(by=['weighted_rating'])['movieId'].unique()
fin = users.drop_duplicates(subset=['movieId'], keep=False).sort_values(by=['weighted_rating']).iloc[::-1]
titles = []
for id in fin['movieId']:
    tle = df[df['movieId']==id]['title'].values[0]
    titles.append(tle)

x_values = fin['weighted_rating'].iloc[:top_n_movies]
y_values = titles[:top_n_movies][::-1]
y_axis   = np.arange(1, top_n_movies+1, 1)
plt.barh(y_axis, x_values, align='center')
plt.yticks(y_axis, y_values)
plt.show()
```
